## Exercise 2 - Ask user for age
`input()` gives text, then we print it in a sentence.

```python
age = input("How old are you? ")
print("You are", age, "years old.")
```

## Exercise 3 - Simple calculator
Convert user input to `int`, then add the numbers.

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("Sum =", num1 + num2)
```

## Exercise 4 - String concatenation
Join strings with `+` and include a space between names.

```python
first = "Alex"
last = "Brown"
full_name = first + " " + last
print(full_name)
```
