## What is a variable?
A variable is a labeled box that stores information.

```python
name = "Mila"
age = 10
print(name)
print(age)
```

## Naming rules
- Use letters, numbers, underscore
- Do not start with a number
- Use clear names

```python
student_name = "Leo"
favorite_game = "chess"
print(student_name, favorite_game)
```

## Basic types
- `str` for text
- `int` for whole numbers

```python
pet = "dog"
count = 3
print(type(pet))
print(type(count))
```

## input()
Use `input()` to ask the user for text.

```python
user_name = input("What is your name? ")
print("Nice to meet you,", user_name)
```
