## Example - Favorite fruits

```python
fruits = ["apple", "banana", "mango"]
print(fruits[0])
fruits.append("orange")
print(fruits)
```

## Example - Loop list

```python
pets = ["dog", "cat", "fish"]
for pet in pets:
    print("I like", pet)
```
