## Exercise 1 - Create a snacks listSolution idea: Create a list with 3 snacks and print it.

## Exercise 3 - Add and removeSolution idea: Add one game to a list, then remove one game.

```python
games = ["chess", "soccer", "tennis"]games.append("coding")games.remove("soccer")print(games)
```
