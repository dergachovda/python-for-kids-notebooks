## Exercise 1 - Build a profileSolution idea: Create a dictionary with name, age, and city, then print it.

```python
profile = {"name": "Alex", "age": 10, "city": "Kyiv"}print(profile)
```

## Exercise 2 - Update valueSolution idea: Change the age value inside a dictionary.

```python
profile = {"name": "Alex", "age": 10}profile["age"] = 11print(profile)
```

## Exercise 3 - Create a tupleSolution idea: Create a tuple with 3 numbers and print the second number.

```python
numbers = (8, 3, 5)print(numbers[1])
```

## Exercise 4 - Dictionary loopSolution idea: Loop through dictionary keys and print key + value.

```python
data = {"math": 95, "science": 98}for key in data:    print(key, data[key])
```
