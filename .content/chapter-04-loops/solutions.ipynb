## Exercise 1 - Print 1 to 10Solution idea: Use a for loop to print numbers from 1 to 10.

```python
for n in range(1, 11):    print(n)
```

## Exercise 2 - Even numbersSolution idea: Print even numbers from 2 to 20.

```python
for n in range(2, 21, 2):    print(n)
```

## Exercise 3 - CountdownSolution idea: Make a countdown from 5 to 1, then print "Go!".

```python
n = 5while n >= 1:    print(n)    n -= 1print("Go!")
```

## Exercise 4 - Stop at 7Solution idea: Loop from 1 to 10 and stop when number is 7.

```python
for n in range(1, 11):    if n == 7:        break    print(n)
```
