## Exercise 1 — Count 1 to 10Task: Print the numbers from 1 to 10.

```python
for n in range(1, 11):    print(n)
```

## Exercise 2 — Even numbersTask: Print even numbers from 2 to 20.

```python
for n in range(2, 21, 2):    print(n)
```

## Exercise 3 — CountdownTask: Count down from 5 to 1, then print Go!

```python
n = 5while n >= 1:    print(n)    n -= 1print("Go!")
```

## Exercise 4 — Stop earlyTask: Print numbers 1 to 10 but stop at 7.

```python
for n in range(1, 11):    if n == 7:        break    print(n)
```
