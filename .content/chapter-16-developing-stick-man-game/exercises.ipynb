## Exercise 1
Implement a `Platform` class using rectangle bounds.

```python
class Platform:
    def __init__(self, canvas, x1, y1, x2, y2):
        self.canvas = canvas
        self.id = canvas.create_rectangle(x1, y1, x2, y2, fill="#4caf50")

    def bounds(self):
        return self.canvas.coords(self.id)
```
