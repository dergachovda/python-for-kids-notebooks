## Exercise 1
Place a door sprite and store its bounds for collision checks.

```python
import tkinter as tk

window = tk.Tk()
canvas = tk.Canvas(window, width=640, height=360, bg="#1f1f1f")
canvas.pack()

# Example with rectangle placeholder
door_id = canvas.create_rectangle(560, 120, 610, 200, fill="gold")

def door_bounds():
    return canvas.coords(door_id)
```

## Exercise 2
Detect overlap between player and door and trigger win.

```python
has_won = False

def check_win(player_box, door_box):
    px1, py1, px2, py2 = player_box
    dx1, dy1, dx2, dy2 = door_box
    return px1 < dx2 and px2 > dx1 and py1 < dy2 and py2 > dy1
```
