## Exercise 1 — Pass checkTask: Ask for a score. Print Pass if score >= 60, else print Retry.

```python
score = int(input("Score: "))if score >= 60:    print("Pass")else:    print("Retry")
```

## Exercise 2 — Color checkTask: Ask for a color. If the color is blue, print a special message.

```python
color = input("Color: ").lower()if color == "blue":    print("Blue is awesome!")else:    print("Nice choice!")
```

## Exercise 3 — Temperature adviceTask: Ask for the temperature and suggest what to wear.

```python
temp = int(input("Temperature: "))if temp > 25:    print("Wear a t-shirt")else:    print("Wear a jacket")
```

## Exercise 4 — Secret wordTask: Print "Correct" only if the user types python.

```python
word = input("Secret word: ")if word == "python":    print("Correct")else:    print("Try again")
```
