## Exercise 1 - Pass or retrySolution idea: Ask for a test score and print "Pass" if score >= 60, else "Retry".

```python
score = int(input("Enter your score: "))if score >= 60:    print("Pass")else:    print("Retry")
```

## Exercise 2 - Favorite color checkerSolution idea: Ask for a color. If color is "blue", print a special message.

```python
color = input("Enter a color: ")if color == "blue":    print("Blue is awesome!")else:    print("Nice choice!")
```

## Exercise 3 - Temperature adviceSolution idea: Ask for temperature. If above 25 print "Wear a t-shirt", otherwise "Wear a jacket".

```python
temp = int(input("Temperature: "))if temp > 25:    print("Wear a t-shirt")else:    print("Wear a jacket")
```

## Exercise 4 - Secret wordSolution idea: Ask for a secret word and print "Correct" only when word equals "python".
