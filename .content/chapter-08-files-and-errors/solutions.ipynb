## Exercise 1 — Write storyTask: Write one sentence to story.txt.

```python
with open("story.txt", "w", encoding="utf-8") as f:    f.write("Today I learned loops!\n")print("Saved")
```

## Exercise 2 — Read storyTask: Read story.txt and print the text.

```python
with open("story.txt", "r", encoding="utf-8") as f:    print(f.read())
```

## Exercise 3 — Append lineTask: Add one more line to story.txt.

```python
with open("story.txt", "a", encoding="utf-8") as f:    f.write("Python is fun!\n")print("Line added")
```

## Exercise 4 — Safe doubleTask: Ask for a number and print number * 2 — handle bad input.

```python
try:    n = int(input("Number: "))    print(n * 2)except ValueError:    print("Please type digits only")
```
