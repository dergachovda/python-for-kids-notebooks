## Exercise 1 - Save one lineSolution idea: Write one sentence into a file named story.txt.

```python
with open("story.txt", "w", encoding="utf-8") as f:    f.write("Today I learned loops!\n")print("Saved!")
```

## Exercise 2 - Read fileSolution idea: Read story.txt and print its content.

```python
with open("story.txt", "r", encoding="utf-8") as f:    text = f.read()print(text)
```

## Exercise 3 - Append lineSolution idea: Add a second line to story.txt.

```python
with open("story.txt", "a", encoding="utf-8") as f:    f.write("Python is fun!\n")print("Line added")
```

## Exercise 4 - Handle bad inputSolution idea: Ask for a number and handle invalid input.

```python
try:    number = int(input("Number: "))    print(number * 2)except ValueError:    print("Please enter digits only.")
```
