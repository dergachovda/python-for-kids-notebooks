## Example - Write and read

```python
with open("notes.txt", "w", encoding="utf-8") as f:    f.write("I love Python\n")with open("notes.txt", "r", encoding="utf-8") as f:    print(f.read())
```

## Example - Safe number input

```python
try:    value = int(input("Enter a number: "))    print("You typed", value)except ValueError:    print("That was not a valid number.")
```
